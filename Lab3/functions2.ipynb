```python
movies = [
{
"name": "Usual Suspects", 
"imdb": 7.0,
"category": "Thriller"
},
{
"name": "Hitman",
"imdb": 6.3,
"category": "Action"
},
{
"name": "Dark Knight",
"imdb": 9.0,
"category": "Adventure"
},
{
"name": "The Help",
"imdb": 8.0,
"category": "Drama"
},
{
"name": "The Choice",
"imdb": 6.2,
"category": "Romance"
},
{
"name": "Colonia",
"imdb": 7.4,
"category": "Romance"
},
{
"name": "Love",
"imdb": 6.0,
"category": "Romance"
},
{
"name": "Bride Wars",
"imdb": 5.4,
"category": "Romance"
},
{
"name": "AlphaJet",
"imdb": 3.2,
"category": "War"
},
{
"name": "Ringing Crime",
"imdb": 4.0,
"category": "Crime"
},
{
"name": "Joking muck",
"imdb": 7.2,
"category": "Comedy"
},
{
"name": "What is the name",
"imdb": 9.2,
"category": "Suspense"
},
{
"name": "Detective",
"imdb": 7.0,
"category": "Suspense"
},
{
"name": "Exam",
"imdb": 4.2,
"category": "Thriller"
},
{
"name": "We Two",
"imdb": 7.2,
"category": "Romance"
}
]

def score_checker():
    movie_name = input("Input movie name:")
    for movie in movies:
        if movie['name'] == movie_name:
            if movie['imdb'] > 5.5:
                return True
            
def score_above():
    res = []
    for movie in movies:
        if movie['imdb'] > 5.5:
            res.append(movie['name'])
    return res

def category_printer():
    category_name = input("Input category of the film:")
    arr = []
    for movie in movies:
        if movie['category'] == category_name:
            arr.append(movie['name'])
    return arr

def imdb_avg():
    film_list = []
    n = int(input("Input number of films"))
    for i in range(n):
        name = input("Enter film name")
        film_list.append(name)
    s = 0
    c = 0
    for movie in movies:
        if movie['name'] in film_list:
            s += movie['imdb']
            c += 1
    return s / c

def imdb_avg_by_category():
    genre_list = []
    n = int(input("Input number of genres"))
    for i in range(n):
        name = input("Enter category name")
        genre_list.append(name)
    s = 0
    c = 0
    for movie in movies:
        if movie['category'] in genre_list:
            s += movie['imdb']
            c += 1
    return s / c

```
